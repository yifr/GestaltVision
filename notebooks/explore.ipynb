```python
import torch
import os
import cv2
import h5py as h5
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calc_mean(dataset):
    mean = np.zeros((3, 256, 256))
    n = 0
    for i in range(dataset.shape[0]):
        img = dataset[i]
        mean += dataset
        n += 1
    mean /= n
    return mean
```
